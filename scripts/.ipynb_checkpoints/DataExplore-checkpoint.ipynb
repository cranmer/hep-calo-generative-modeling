# Data

#### We will be working with a sample dataset of 100 experiments, each experiment has 3 data components:  
ECAL: (24,24,25) array of energy reading  
HCAL: (4,4,60) array of XXX readings  
target: (5) array specifying [partice_type, momentum_X, momentum_Y, XXX, XXX]

```python
data = data_loader.load_test_data()

print 'data.keys: ',data.keys()
ecal = data['ECAL'][:]
hcal = data['HCAL'][:]
target = data['target'][:]
particle_type = target[:,:,1].ravel()

print 'eacl: ', ecal.shape
print 'hcal: ', hcal.shape
print 'target', target.shape
print 'particle', particle_type.shape
```

### Investigate 5 subsets of the data:
1) a single particle=='proton' experiment  
2) a single particle=='electron' experiment  
3) the average across all of the particle=='proton' experiments  
4) the average across all of the particle=='electron' experiments 
5) the average across all experiments

```python
single_p = ecal[0]
single_e = ecal[1]
# NB: the three below are all the same, this is because the test-sample-data is coming from 'Gamma' and not 'shuffled'
mean_p = ecal.mean(axis=0) 
mean_e = ecal.mean(axis=0)
mean_all = ecal.mean(axis=0)
```

### Let's take a look at a 3D-scatterplot of the shower to get an idea
 
Signature: viz_tools.scatter3d(frame, scaling_root=1, top=None, save=None)  
Docstring:  3D scatterplot with colormap for energy values  
input:  
>  
scaling_root: data**(1/scaling_root) for better colorbar visualization  [data skew]  
top: keep top% highest energy values [data skew]  
save: path to save plot  

Examples illustrating how to use the ***top*** and ***scaling_root*** arguments shown below

##### Default usage

```python
ecal_df = viz_tools.np2df(single_p)
viz_tools.scatter3d(ecal_df)
```

##### Only keeping the top 40% of energy values
Allows to focus the plot on the high energy values

```python
viz_tools.scatter3d(ecal_df, top=40)
```

##### Rescaling the energy values
This way the colormap is not dominated by the few very high energy points

```python
viz_tools.scatter3d(ecal_df, top=25, scaling_root=5)
```

### A note on scaling_factor and binary option in visualizations
All of the other visualization functions have the scaling_factor and even a binary option.  
The binary version only plots non-zero energy points (examples below)

# Slice Plots
These plots take a 2D slice from the 3D shower-tensor  
If slice index locations (a list) is not specified, a series of plot "walking through" the entire tensor will be generated

#### Slice into the 14th index location of the Z-dimension
- a default plot version
- a version with the scaling factor set to 5 (note how the scale on the legend changes)
- a version with binary==True (any non-zero value is plotted)

#### For one particular experiment

```python
viz_tools.slice_axis_heatmap(single_p, 2, location=14)
viz_tools.slice_axis_heatmap(single_p, 2, location=14, scaling_root=5)
viz_tools.slice_axis_heatmap(single_p, 2, location=14, binary=True)
```

#### For the 'average' experiment


```python
viz_tools.slice_axis_heatmap(mean_p, 2, location=14)
viz_tools.slice_axis_heatmap(mean_p, 2, location=14, scaling_root=5)
viz_tools.slice_axis_heatmap(mean_p, 2, location=14, binary=True)
```

Let's see what happend when no index location is passed and a series of plots "walking through" the Z-axis are displayed

```python
viz_tools.slice_axis_heatmap(mean_p, 2, scaling_root=2)
```

### Projecting to 2D by taking the mean 

For the 'average' experiment

```python
viz_tools.collapse_axis_heatmap(mean_p, 0, scaling_root=3)
viz_tools.collapse_axis_heatmap(mean_p, 1, scaling_root=3)
viz_tools.collapse_axis_heatmap(mean_p, 2, scaling_root=3)
```

For a single experiment

```python
viz_tools.collapse_axis_heatmap(single_p, 0, scaling_root=3)
viz_tools.collapse_axis_heatmap(single_p, 1, scaling_root=3)
viz_tools.collapse_axis_heatmap(single_p, 2, scaling_root=3)
```

### Axis Signature
We can take the idea of collapsing an axis one step further and take an average across 2 axes to generate an axis specific 'signature'.   
Notice the distributional stability we obtain by looking at the average-experiment rather than a single trail.

X-dimension

```python
viz_tools.hist_by_dim(single_p,0)
viz_tools.hist_by_dim(mean_p,0)
```

Y-dimension

```python
viz_tools.hist_by_dim(single_p,1)
viz_tools.hist_by_dim(mean_p,1)
```

Z-dimension

```python
viz_tools.hist_by_dim(single_p,2)
viz_tools.hist_by_dim(mean_p,2)
```
